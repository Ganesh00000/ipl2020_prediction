```python
ipl_teams=['Sunrisers Hyderabad', 'Mumbai Indians', 
        'Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals',
         'Delhi Capitals']
```

```python
matches_=matches[(matches["team1"].isin(ipl_teams))&(matches["team2"].isin(ipl_teams))]
```

```python
matches_.head()
```

```python
matches_.shape
```

```python
df_=matches_[["team1","team2","toss_winner","winner"]]
```

```python
df_.head()
```

```python
df_.isnull().sum()
```

```python
df_.dropna(how="any",inplace=True)
```

```python
df_copy=df_.copy()
```

```python
final=pd.get_dummies(df_,prefix=["team1","team2"],columns=["team1","team2"])
```

```python
final.head()
```

```python
x=final.drop(["winner","toss_winner"],axis=1)
```

```python
y=final["winner"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
regressor=LogisticRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_predict=regressor.predict(x_test)
```

```python
print(pd.DataFrame({"Actual":y_test,"Predicted":y_predict}))
```

```python
from sklearn import metrics
```

```python
print(metrics.accuracy_score(y_test,y_predict))
```
